```python
import os
import re
import glob
import numpy as np
import xarray as xr

xr.set_options(display_expand_data=False)

%config InlineBackend.figure_format='retina'
```

```python
#TSIDs:
TSIDs = ['TEST-HOP', 'VST']

#Data File Path:
file_path = os.path.join(glob.glob('C:\\Users\\mvomstein\\projects\\vt-2pod')[0], 'DATA')

file_path_save = glob.glob(
    'C:\\Users\\mvomstein\\projects\\vt_2pod_data_analysis\\DATA_xr')[0]
```

```python
def getMatch(param, fpts, bs, hv):
    matches = [match for match in os.listdir(fpts) 
           if bs in match 
           and hv in match 
           and param in match
           and os.path.splitext(match)[1] == '.npy'
          ]
    if len(matches) >=2: raise Exception('MULTIPLE ENTRIES: DATA NOT CLEAN')
    return(matches)
```

```python
BSs = ['forearm', 'back-lower', 'abdomen', 'thigh']
HVs = ['_h_', '_v_']
bsDic = {'forearm': 0,
         'back-lower': 1,
         'abdomen': 2,
         'thigh': 3}

hvDic = {'h': 0,
         'v': 1}

params = ['postmean']
stimRange = np.arange(2.5, 61, 2.5)

for TSID in TSIDs:
    fpTS = os.path.join(file_path, TSID)
    arr = np.random.normal(size=(24,1,4,2))*np.nan
    for BS in BSs:
        for HV in HVs:
            matches = getMatch('postmean', fpTS, BS, HV)
            print(matches)
            if matches:
                x = re.split("_", matches[0])
                singleData =np.load(
                    os.path.join(fpTS, matches[0]), allow_pickle=True)
                arr[:len(singleData),0, bsDic[x[1]], hvDic[x[2]]] = singleData
                
                #x[0]: TSID, x[1]: bs, x[2]: hv
                
    postmean = xr.DataArray(
        arr,
        dims=['stimRange', 'tsid', 'bs', 'hv'],
        coords=dict(stimRange=stimRange, tsid=[TSID], bs=BSs, hv=['h', 'v']),
        name='postmean'
    )
    postmean.to_netcdf(os.path.join(file_path_save, 'postmean', 'postmean'+TSID))
```
